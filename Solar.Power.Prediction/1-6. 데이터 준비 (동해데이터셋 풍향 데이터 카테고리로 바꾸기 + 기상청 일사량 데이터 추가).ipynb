### 1) 풍향 각도 데이터 카테고리형태로 변경
- 기존 : "wind_angle" column, 1~360
- 추가 : "wind_direction" column, 1~8로 변경 (270도부터 1, 45도 간격)

```python
import pandas as pd
```

```python
df = pd.read_csv('donghae_solar_power_dataset2018_rev3.csv', index_col=0)
```

```python
df.head()
```

```python
df['wind_direction'] = 0
```

```python
df.head()
```

```python
# 각도별 방향값 
dir1 = df[(df['wind_angle'] >= 270) & (df['wind_angle'] < 315)].index
dir2 = df[(df['wind_angle'] >= 315) & (df['wind_angle'] <= 360)].index
dir3 = df[(df['wind_angle'] >= 0) & (df['wind_angle'] < 45)].index
dir4 = df[(df['wind_angle'] >= 45) & (df['wind_angle'] < 90)].index
dir5 = df[(df['wind_angle'] >= 90) & (df['wind_angle'] < 135)].index
dir6 = df[(df['wind_angle'] >= 135) & (df['wind_angle'] < 180)].index
dir7 = df[(df['wind_angle'] >= 180) & (df['wind_angle'] < 225)].index
dir8 = df[(df['wind_angle'] >= 225) & (df['wind_angle'] < 270)].index
```

```python
df.iloc[dir1, -1] = 1
df.iloc[dir2, -1] = 2
df.iloc[dir3, -1] = 3
df.iloc[dir4, -1] = 4
df.iloc[dir5, -1] = 5
df.iloc[dir6, -1] = 6
df.iloc[dir7, -1] = 7
df.iloc[dir8, -1] = 8
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[['date', 'hour', 'humidity', 'rain', 'rain_type', 'temp', 'thunder',
       'wind_angle', 'wind_direction', 'wind_velo', 'sky_con', 'solar_radiation_normal',
       'season', 'SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM25', 'power',
       'date_time']]
```

```python
df.head()
```

```python
df.wind_direction.unique()
```

### 2) 기상청 일사량 데이터(강릉) 추가

```python
import pandas as pd
```

```python
df = pd.read_csv('donghae_solar_power_dataset2018_rev5.csv', index_col=0)
```

```python
df2 = pd.read_csv('동해기상청일사량데이터.csv')
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2.columns
```

```python
df2 = df2[['일시', '일사(MJ/m2)']]
```

```python
df2.columns = ['date', 'solar_radiation']
```

```python
df2.head()
```

```python
df2['time'] = pd.to_datetime(df2['date'])
```

```python
df2.head()
```

```python
df2['time_str'] = df2['time'].astype(str)
```

```python
df2.head()
```

```python
# 시간 칼럼 만들기 
hours = []
for i in range(len(df2.date)):
    hour = df2['time_str'][i][11:13]
    hours.append(hour)
df2['hour'] = hours
```

```python
# 0없애기
df2.hour=df2.hour.astype(int)
df2.hour=df2.hour.astype(str)
```

```python
df2.head()
```

```python
df2['date_str'] = 0
for i in range(len(df2.date)):
    df2['date_str'][i] = df2['time_str'][i].split(' ')[0]
```

```python
df2.tail()
```

```python
df2.hour.unique()
```

```python
# 키값 만들기
df2['date_time'] = df2['date_str'] + '-' + df2['hour']
df2.head()
```

```python
df3 = df2[['date_time', 'date_str','hour','solar_radiation']]
```

```python
df3.head()
```

```python
df3 = df3[['date_time', 'solar_radiation']]
```

```python
df3.head()
```

```python
df4 = pd.merge(df,df3,how='outer')
```

```python
df4=df4.fillna(0)
```

```python
df4.isnull().sum()
```
