# 동해시 미세먼지 데이터 만들기

- 데이터 url : https://www.airkorea.or.kr/web/last_amb_hour_data?pMENU_NO=123
    - 에어코리아 > 통계정보 > 최종확정 측정자료 조회 > 확정자료 다운로드 
- 데이터 : 최종확정자료 연도별 다운로드 후 엑셀로 동해 천곡동만 빼내서 저장 

- 작업내용 
    - "측정일시" column 2018010101 연 월 일 시간 붙어 있는 형태를 연-월-일-시간 형태로 변경
    - 6~20시만 남기고 삭제 
    
    - 앞의 과정과 거의 유사하므로 상세 주석 생략함

```python
# 라이브러리 로딩
import pandas as pd
```

```python
# 데이터 로딩
df=pd.read_excel('동해대기데이터.xlsx')
```

```python
df.head()
```

```python
df.columns
```

```python
# 필요한 column만 선택
df = df[['측정일시', 'SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM25']]
```

```python
df.head()
```

```python
df.tail()
```

```python
# 키값 만들기 형태 : 2018-01-01-01
df['time'] = df['측정일시'].astype(str)
```

```python
times = []
for i in range(len(df['측정일시'])):
    time = df['time'][i][8:]
    times.append(time)
    
df['hour'] = times
```

```python
df['hour_int'] = df['hour'].astype(int)
df['hour_int'] = df['hour_int'].astype(str)
```

```python
df.head()
```

```python
# 20180101 분리
dates = []
for i in range(len(df['측정일시'])):
    date = df['time'][i][:8]
    dates.append(date)
    
df['date'] = dates
```

```python
df.head()
```

```python
# 2018-01-01 형태 만들기
import datetime
date_times=[]
for i in range(len(df['date'])):
    date_time = datetime.datetime.strptime(df['date'][i], '%Y%m%d')
    date_times.append(date_time)
    
df['date_time'] = date_times

df['date_time'] = df['date_time'].astype(str)

df.head()
```

```python
# 2018-01-01-1 형태 만들기
df['date_time2'] = df['date_time'] + '-' + df['hour_int']
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns
```

```python
df100 = df[['date_time', 'hour_int', 'SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM25','date_time2']]
```

```python
df100.head()
```

```python
df100.tail()
```

```python
df100.columns = ['date', 'hour', 'SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM25', 'date_time']
```

```python
df100.head()
```

```python
df100
```

```python
df100['hour']=df100['hour'].astype(int)
```

```python
# 해 없는 시간 제거하기 
result = []
rej_time = [1,2,3,4,5, 21,22,23,24]
for j in rej_time:
    one_indice = []
    for i in range(len(df100[df100['hour'] ==j].index)):
        one = df100[df100['hour'] ==j].index[i]
        one_indice.append(one)
        
    result.extend(one_indice)    

result

df100.drop(result, inplace=True)
df100.reset_index(drop=True, inplace=True)

df100.head()
```
