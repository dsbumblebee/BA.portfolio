## 1) 기상데이터 불러오기 

작업내용 : 데이터 불러온 후 키값 생성 

```python
import pandas as pd
```

```python
weather = pd.read_csv('donghae_weather_2018.csv', index_col=0)
```

```python
weather.head()
```

```python
# merge 위해서 key값 만들기 위해 date_time 칼럼 생성 
# 연 - 월 - 일 형식을 맞추기 위함, 2018-06-01  , 2018-6-1 다르기 때문에 datetime.strptime 변경 후 astype으로 다시변경 

import datetime

date_times=[]
for i in range(len(weather['date'])):
    date_time = datetime.datetime.strptime(weather['date'][i], '%Y-%m-%d')
    date_times.append(date_time)
    
weather['date_time'] = date_times
weather['date_time'] = weather['date_time'].astype(str)
```

```python
weather.head()
```

```python
weather.drop('date', axis=1, inplace=True)
weather['hour']= weather['hour'].astype(str)
weather.columns = ['hour', 'humidity', 'rain', 'rain_type', 'temp', 'thunder',
       'wind_angle', 'wind_velo', 'sky_con', 'date']
weather.head()
```

```python
# 키값 만들기
# merge위해서 년 - 월 - 일 - 시간 형식으로 변경 후 'hour', 'date' column 제거 
weather['date_time'] = weather['date'] + '-' +weather['hour']
weather2 = weather.drop(['hour','date'], axis=1)
weather2.head()
```

## 2) 발전량 데이터 불러오기

데이터 불러온 후 키값 생성 

```python
power = pd.read_csv('donghae_solar_power2018.csv', index_col=0)
```

```python
power.head()
```

```python
# column 선택
power = power[['date','hour','power']]
```

```python
power.head()
```

```python
power.tail()
```

```python
# 해 없는 시간 제거하기 
# 6~20시만 사용할 예정이기 때문에, 1,2,3,4,5, 21,22,23,24 이 포함된 index를 추출함

# 작업방식 : 필요없는 시간의 index를 추출하여 drop

result = []
rej_time = [1,2,3,4,5, 21,22,23,24]
for j in rej_time:
    one_indice = []
    for i in range(len(power[power['hour'] ==j].index)):
        one = power[power['hour'] ==j].index[i]
        one_indice.append(one)
        
    result.extend(one_indice)    

result
```

```python
# 추출한 index drop
power.drop(result, inplace=True)
power.reset_index(drop=True, inplace=True)
```

```python
power.head()
```

```python
# merge 위해서 key값 만들기 위해 시간 string으로 변경 후 "date" colume과 결합
power['hour']=power['hour'].astype(str)
power['date_time'] = power['date'] + '-' + power['hour']
```

```python
power.head()
```

```python
power = power[['date_time','power']]
power.head()
```

## 3) 1),2) 데이터 결합하기

```python
# merge 함수로 데이터 결합하기
donghae_power_dataset = pd.merge(weather2, power)
```

```python
donghae_power_dataset.head()
```

```python
# hour 추가
donghae_power_dataset['hour'] = weather['hour']
```

```python
# date 추가 
donghae_power_dataset['date'] = weather['date']
```

```python
# 잘 결합 되었는지 확인
donghae_power_dataset.iloc[45:90]
```

```python
donghae_power_dataset.columns
```

```python
# column 순서 정리
donghae_power_dataset = donghae_power_dataset[['date', 'hour', 'humidity', 'rain', 'rain_type', 'temp', 'thunder', 'wind_angle',
       'wind_velo', 'sky_con', 'power', 'date_time']]
```
