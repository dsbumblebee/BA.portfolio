## 2-3. 기초 통계량 

```python
def func(x):
    d = {}
    d['Count'] = round(x['제목'].count())
    d['AverageRating'] = round(x['별점'].mean(),3)
    d['SDrating'] = round(x['별점'].std(),3)
    d['NegRate'] = round(x['v_neg'].mean(),3)
    d['PosRate'] = round(x['v_pos'].mean(),3)
    d['Price'] = round(x['가격'].mean(),0)
    return pd.Series(d, index=['Count','AverageRating','SDrating','NegRate','PosRate','Price'])
```

```python
data.groupby(['카테고리_theo','카테고리']).apply(func)
```

## 3. 순서형 프로빗 모형

```python
# 데이터 로딩
data = pd.read_csv('frip_data_result_ver2.csv', index_col=0)

# 공부, 재테크 여가의 범주에 없음 
data = data[(data['카테고리']!='공부')&(data['카테고리']!='재테크')]

data['평점_cat_1unit'] = data.별점.replace([5. , 4.5, 4. , 3.5, 3. , 2.5, 2. , 1.5, 1. , 0.5],
               [5,4,4,3,3,2,2,1,1,0])
data['평점_cat_1unit'] = data['평점_cat_1unit'].astype('category')
```

```python
# 필요한 데이터만 
data = data[['별점','v_neg','v_pos','Topic_code','카테고리_theo','가격','평점_cat_1unit']]

# 가격 데이터 추가 (만원단위)
data['가격_만원'] = data.가격/10000
```

```python
# 데이터 더미화
topic_dum = pd.get_dummies(data['Topic_code'])
cat_dum = pd.get_dummies(data['카테고리_theo'])

data = pd.concat([data,topic_dum,cat_dum], axis=1)

# 더미데이터 중 1개씩 제외, 더미데이터의 원 데이터 제외 
data.drop(['Topic_code','사회적', '카테고리_theo','기타'], axis=1, inplace=True)
```

```python
# 모델 생성
from statsmodels.miscmodels.ordinal_model import OrderedModel

mod_prob = OrderedModel(data['평점_cat_1unit'], data[['v_neg', 'v_pos', '심리적', '신체적', '개인적만족', '자기계발', '사교적', '스포츠',
       '야외활동', '예술활동', '가격_만원']], distr='probit')
```

```python
# 모델 결과 
res_prob = mod_prob.fit(method='bfgs')
res_prob.summary()
```
